```python
file_paths = {
    'RELAINCE': '/content/RELIANCE.NS.csv',
    'TATA': '/content/TATAPOWER.NS.csv',
    'ADANI': '/content/ADANIPOWER.NS.csv',
}
data = pd.DataFrame()

for stock, file_path in file_paths.items():
    stock_data = pd.read_csv(file_path, parse_dates=['Date'])
    data[stock] = stock_data['Close']

daily_returns = data.pct_change()

cumulative_returns = (1 + daily_returns).cumprod() - 1

plt.figure(figsize=(14, 8))
for stock in file_paths.keys():
    plt.plot(cumulative_returns.index, cumulative_returns[stock], label=stock)

plt.title('Cumulative Returns of Multiple Stocks', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Cumulative Returns', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import yfinance as yf

stocks = ["ADANIPOWER.NS", "TATAPOWER.NS", "NTPC.NS", "RPOWER.NS"]
stocks_data = pd.DataFrame()

for stock in stocks:
    data = yf.download(stock, start='2014-01-01', end='2023-12-31')['Close']
    stocks_data[stock] = data

returns = stocks_data.pct_change()
returns_cleaned = returns.dropna()

annual_returns = (1 + returns_cleaned).resample('Y').prod() - 1

print("Annual Returns for Each Stock:")
print(annual_returns)
```
